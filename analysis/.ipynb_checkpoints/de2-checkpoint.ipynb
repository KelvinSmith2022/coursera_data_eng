```python
def convert_timestamp(df, col):
    df[col+'_Year'] = df[col].dt.year
    df[col+'_Month'] = df[col].dt.month
    df[col+'_Day'] = df[col].dt.day
    df.drop(col, axis=1, inplace=True)
```

```python
convert_timestamp(data, 'timestamp')
```

```python
data_unique(data, 'price')
```

```python
def mean_encoding(df, col1, col2):
    dict = df.groupby(col1)[col2].mean().to_dict()
    df[col1] = df[col1].map(dict)
```

```python
# applying feature selection technique
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso
```

```python
y = data['price']
x = data.drop('price', axis=1)
```

```python
feat_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
```

```python
feat_model.fit(x,y)
```

```python
feat_model.get_support()
```

```python
cols = x.columns
```

```python
sel_feat = cols[feat_model.get_support()]
```

```python
x = x[sel_feat]
x
```

```python
# applying model selection
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model1 = KNeighborsClassifier(n_neighbors=3)
```

```python
model1.fit(X_train, y_train)
```

```python
prediction = model1.predict(X_test)
```

```python
print(prediction)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(prediction, y_test)
```

```python
confusion_matrix(prediction, y_test)
```
